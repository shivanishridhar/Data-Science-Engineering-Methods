```python
#Importing Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import ShuffleSplit
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
from tensorflow import keras 
from keras.layers import Dense
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from sklearn.model_selection import RandomizedSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
```

```python
loan_dataset = pd.read_excel('/content/loan.xlsx')
loan_dataset
```

```python
#dropping irrelevant column
loan_dataset.drop('Telephone',axis=1,inplace=True)
loan_dataset.head()
```

```python
loan_dataset['Decision'] = loan_dataset['Decision'].replace(['accept','reject'],[1,0])
```

```python
#dropping irrelevant column
loan_dataset.drop('Sex',axis=1,inplace=True)
loan_dataset.head()
```

```python
#columns of data that will be used for classification as X variable
X_train = loan_dataset.drop('Decision',axis=1).copy()
X_train.head()
```

```python
#column of data that we need to predict as Y variable
Y_train = loan_dataset['Decision'].copy()
Y_train.head()
```

```python
#numerical columns
numerical_input = X_train.select_dtypes(include="number")
numerical_input.columns
```

```python
#categorical columns
categorical_input = X_train.select_dtypes(include='object')
categorical_input.columns
```

```python
from sklearn.preprocessing import scale
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing
```

```python
len(loan_dataset)
```

```python
le = preprocessing.LabelEncoder()
```

```python
revised_categorical_input = categorical_input.apply(le.fit_transform)
revised_categorical_input
```

```python
#concatenating numerical dataset & label encoded categorical dataset
revised_input = pd.concat(objs= [revised_categorical_input,numerical_input],axis='columns')
revised_input.head()
```

```python
#OneHotEncoder
enc = preprocessing.OneHotEncoder()

#fit
enc.fit(categorical_input)

# transform
onehotlabel = enc.transform(categorical_input).toarray()
onehotlabel.shape
```

```python
onehotlabel
```

```python
loan_dataset.isnull().sum()
```

```python
revised_input
```

```python
X_train, X_test, y_train, y_test = train_test_split(onehotlabel, Y_train, test_size=0.2)
```

```python
# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
# Define the model architecture
model = Sequential()
model.add(Dense(8, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy','mse', 'mae'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=16, verbose=1)
```

```python
# Evaluate the model on the testing set
score = model.evaluate(X_test, y_test, verbose=0)
print(f'Test loss: {score[0]}')
print(f'Test accuracy: {score[1]}')
print(f'mse: {score[2]}')
print(f'mae: {score[3]}')
```

```python
param_dist = {
    'hidden_layer_sizes': [(8, 4), (16, 8)],
    'activation': ['relu', 'sigmoid'],
    'optimizer': ['adam', 'sgd'],
    'batch_size': [16, 32, 64],
    'epochs': [50, 100, 200]
}
```

```python
# Define the model architecture
def create_model(hidden_layer_sizes=(8, 4), activation='relu', optimizer='adam'):
    model = Sequential()
    model.add(Dense(hidden_layer_sizes[0], input_dim=X_train.shape[1], activation=activation))
    model.add(Dense(hidden_layer_sizes[1], activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
keras_classifier = KerasClassifier(build_fn=create_model, verbose=0)
```

```python
# Perform randomized search to find the best hyperparameters
random_search = RandomizedSearchCV(keras_classifier, param_distributions=param_dist, cv=3, n_iter=10, n_jobs=-1)
random_search.fit(X_train, y_train)
```

```python
# Print the best hyperparameters and accuracy score
print(f'Best hyperparameters: {random_search.best_params_}')
print(f'Accuracy score: {random_search.best_score_:.4f}')

score_hyperparameter_dense=random_search.best_score_
```
