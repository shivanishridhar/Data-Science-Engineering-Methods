```python
#1.c.Reverse a list
def reverse_list(elements,n):

	if(n==1):
		return elements
	

	elif(n==2):
		elements[0],elements[1],=elements[1],elements[0]
		return elements
	

	else:
		i=0
		while(i<n//2):


			elements[i],elements[n-i-1]=elements[n-i-1],elements[i]
	

			if((i!=i+1 and n-i-1 != n-i-2) and (i!=n-i-2 and n-i-1!=i+1)):
				elements[i+1],elements[n-i-2]=elements[n-i-2],elements[i+1]
			i+=2
		return elements

elements=[1,2,3,4,5]
n=5
print('Original list: ',elements)
print("Reversed list: ",reverse_list(elements,n))
```

```python
#12.Python program to find the n-th term in a Fibonacci series using recursion. 
def fibo(n):
    if n <= 0:
        print("Incorrect value")
        return np.nan
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
print(fibo(10))
```
